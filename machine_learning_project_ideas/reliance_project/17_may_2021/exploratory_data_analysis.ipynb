```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.size']=20
```

```python
file_name='data_for_reliance_project_02_03_2021_modified.csv'
data=pd.read_csv(file_name)
print(data.shape)
data
```

```python
data=data.dropna(axis=1)
data=data.loc[:, (data != data.iloc[0]).any()] 
print(data.shape)
data
```

```python
y_name='Activity (Kg/g Cat.h)'
y=data[y_name]
def plot_dist(data,column,rotation=0,target_var_plot=False,name='test'):
    print(column)
    print(data[column].nunique())
    print(data[column].value_counts())
    plt.figure(figsize=(16,9))
    plt.hist(data[column],bins=50)
    plt.ylabel('Frequency')
    plt.title(column)
    plt.xticks(rotation=rotation)
    plt.tight_layout()
    plt.savefig(f'output/{name}_dist.jpg')
    if target_var_plot:
        plt.figure(figsize=(16,9))
        plt.scatter(data[column],y)
        plt.ylabel(y_name)
        plt.xlabel(column)
        plt.xticks(rotation=rotation)
        plt.tight_layout()
        plt.savefig(f'output/{name}_trend.jpg')
```

# Distributions

```python
plt.figure(figsize=(16,9))
plt.hist(y,bins=100)
plt.ylabel('Frequency')
plt.xlabel('Activity (Kg/g Cat.h)');
plt.tight_layout()
plt.savefig('output/activity_dist.jpg')
```

```python
column='External Donor'
plot_dist(data,column,rotation=90,target_var_plot=True,name='external_donor')
```

```python
column=' Donor/AlEt3 (External Donor (Properties)'
plot_dist(data,column,target_var_plot=True,name='Mole Fraction')
```

```python
column='Temperature'
plot_dist(data,column,name='temperature')
```
