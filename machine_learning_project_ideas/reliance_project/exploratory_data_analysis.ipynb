```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_name='data_for_reliance_project_02_03_2021_modified.csv'
df=pd.read_csv(file_name)
print(df.shape)
df
```

```python
df=df.dropna(axis=1)
print(df.shape)
df
```

```python
plt.hist(df['Activity (Kg/g Cat.h)'],bins=100)
plt.xlabel('Activity (Kg/g Cat.h)');
```

```python
print('External Donor:\n')
print(df['External Donor'].nunique())
print(df['External Donor'].value_counts())
```
