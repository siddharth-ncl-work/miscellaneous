```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_act=pd.read_csv('CMNPD_1.0_act_br.tsv',sep='\t')
df_act
```

```python
df_act['COMPOUND_ID'].nunique()
```

```python
df_uniq_act=df_act['BIOACTIVITY_IN_BRIEF'].value_counts().to_frame().reset_index().rename(columns={'index':'bioactivity','BIOACTIVITY_IN_BRIEF':'counts'})
df_uniq_act
```

```python
classe_all=list(df_uniq_act['bioactivity'])
classe_all
```

```python
def filterClassCytotoxic(x):
    inclustion_term_list=['cytotox','Cytotox']
    exclusion_term_list=['no cytotox',
                         'No cytotox',
                         'not cytotox',
                         'Not cytotox',
                         'non-cytotox',
                         'Non-cytotox',
                         'non cytotox',
                         'Non cytotox',
                         'noncytotoxic',
                         'Noncytotoxic',
                         'no AB or cytotox. activ',
                         'No AB or cytotox. activ',
                         'No AB, cytotox',
                         'no AB, cytotox',
                         'without notable cytotoxicity',
                         'not AM or cytotox',
                         'Not AM or cytotox',
                         'no significant cytotox',
                         'No significant cytotox',
                         'no AB, AF activities or cytotox',
                         'No AB, AF activities or cytotox'
                        ]
    for inclusion_term in inclustion_term_list:
        if inclusion_term in x:
            for exclution_term in exclusion_term_list:
                if exclution_term in x:
                    return False
            else:
                return True
        else:
            return False
    
class_cytotoxic=list(filter(filterClassCytotoxic,classe_all))
len(class_cytotoxic),class_cytotoxic[-500:]
```

```python
df_act[df_act['BIOACTIVITY_IN_BRIEF'].isin(class_cytotoxic)]
```

```python
def filterClassAM(x):
    inclustion_term_list=['antimicrobial','Antimicrobial','AM']
    exclusion_term_list=[
                        ]
    for inclusion_term in inclustion_term_list:
        if inclusion_term in x:
            for exclution_term in exclusion_term_list:
                if exclution_term in x:
                    return False
            else:
                return True
        else:
            return False
    
class_AM=list(filter(filterClassAM,classe_all))
len(class_AM),class_AM
```

```python
df_act[df_act['BIOACTIVITY_IN_BRIEF'].isin(class_AM)]
```
